合併各類別的 csv，整合進規則中

```python
csv_list = ['device.csv', 'file.csv', 'info_id_time.csv', 'net.csv', 'proc.csv']

df_added = None
for file in csv_list:
    if df_added is None:
        df_added = pd.read_csv(file)
    else:
        df_new = pd.read_csv(file)
        df_added = pd.concat([df_added, df_new], ignore_index=True)

def clean_braces(x):
    x = x.replace("()", "").replace(",", "").replace(" ", "")
    return str(x)
def add_braces(x):
    x = clean_braces(x)
    x = x+"()"
    return x

df_added['Syscall'] = df_added['Syscall'].apply(add_braces)
df_added = df_added.assign(Source='man')
df_added = df_added.reindex(columns=['EntityType','ActionType', 'Source', 'Syscall', 'EnVerb', 'Sentence'])
print(f"num of rule added:{len(df_added)}, \
      num of syscall added: {len(df_added['Syscall'].unique())}, \
      num of verb in new rules: {len(df_added['EnVerb'].unique())}")
df_added.head()
```
