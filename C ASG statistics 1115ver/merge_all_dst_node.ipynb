```python
df = pd.DataFrame(columns=['text','type'])
for typeOfObj in base_data.keys():
    for obj in base_data[typeOfObj]:
        if obj.find('/etc/sed') >= 0:
            continue
        obj = re.sub(r"\"", '', obj)
        if obj.startswith(' ') or obj in ['.', '/']: # 過濾掉' nul'
            continue
        df = df.append({'text': obj, 'type': typeOfObj}, ignore_index=True)
df
```
