```python
import pandas as pd
from collections import Counter

df = pd.read_csv('./驗證實驗_exported.csv')
df = df[['Genre','Pred_obj','Label_obj']]
df = df.dropna()

print(len(df))
df
```

```python
# 連 M, H, L 都一起算有 53 篇
labeled = pd.read_csv('./驗證實驗_exported.csv')
labeled = labeled[['Genre','Label_obj']]
labeled = labeled.dropna()
print(len(labeled))
Counter(labeled['Genre'])
```

```python
from sklearn.metrics import classification_report

y_true = df['Label_obj']
y_pred = df['Pred_obj']
# target_names = ['class 0', 'class 1']
print(classification_report(y_true, y_pred, digits=4)) # target_names=target_names
```

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
(tn, fp, fn, tp)
```

```python
TPR = tp /(tp + fn)
print("True_Positive_Rate:", TPR)
```

- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.cohen_kappa_score.html
- https://www.yongxi-stat.com/cohens-kappa/

```python
from sklearn.metrics import cohen_kappa_score
cohen_kappa_score(y_true, y_pred, labels=['H', 'L'])
# 判斷資料一致性
# > 0.75 優良，0.4 ~ 0.75 一般
```
