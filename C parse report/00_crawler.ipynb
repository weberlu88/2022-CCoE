```python
# Ignore header and footer
filename = "Darlloz giac.org - introduction linux based malware (tech).pdf"
reader = PdfReader(f"report htmls/{filename}")
page = reader.pages[3]

parts = []

def visitor_body(text, cm, tm, fontDict, fontSize):
    y = tm[5]
    if y > 50 and y < 720:
        parts.append(text)

page.extract_text(visitor_text=visitor_body)
text_body = "".join(parts)

print(text_body)
```

```python
# join broken lines
origin_text = """Besides IoT, Zollard also compromised ICS (Industrial Control Systems) [25] devices.
Such as Heat pump, ICS Data Acquisition Server.
Why it can do? The possible reason is, it can launch RCE and dictionary attack.
Even Zollard don’t have ICS devices default password.
From Table 3.
As we can see, the victims include Heat pump, Multichannel Energy Meter, Network Socket and etc..
Some of victim are crossing Internet and LAN (Local Area Network).
Few of them are Spread Node.
It means, it has ability to attack other ICS devices in LAN environment."""

def process(origin_text):
    lines = origin_text.split("\n")
    # for i in range(len(lines)):
    #     lines[i] = lines[i].lstrip()
    oneline_text = "".join(lines)
    result = oneline_text.replace('. ', '.\n')
    new_result = []
    for line in result.split("\n"):
        line = line.lstrip()
        new_result.append( line )
        print(line)

process(origin_text)
# print(new_result)
```

```python
print("total pages", len(reader.pages))
# for page in reader.pages:
#     text = page.extract_text()
#     print(text)
    
for i in range(len(reader.pages)):
    page = reader.pages[i]
    raw_text = page.extract_text()
    raw_text = raw_text.replace('!', ' ')
    raw_text = raw_text.replace('Author Name, email@address', '') # remove footer
    raw_text = raw_text[5:] # remove page num
    res = process(raw_text)
    print(res)
```
