```python
import glob

folder_path = './'
blacklist = ['skip', 'exclude', 'ignore']  # List of strings to exclude from search

# Get a list of all .html files under the folder_path, excluding those that contain a string in the blacklist
# f"{folder_path}/**/*.html"
html_files = [f for f in glob.glob(f"{folder_path}/**/*.html", recursive=True) if not any(black in f for black in blacklist)]
pdf_files  = [f for f in glob.glob(f"{folder_path}/**/*.pdf", recursive=True) if not any(black in f for black in blacklist)]
html_files.extend(pdf_files)
html_files.sort()

print(html_files)
```

```python
def parse(filename:str, familyname):
    if filename.endswith('html'):
        filename = filename.split('.html')[0]
    else:
        filename = filename.split('.pdf')[0]
    filename = ''.join(filename.split(familyname)[1:])
    # print(filename)
    # website, *title = filename.split('-')[0]
    # title = ''.join(title) # bugs
    website = filename.split('-')[0]
    title   = filename.split('-')[1:]
    title = ''.join(title)
    # print(website, title)
    genre = title.split('(')[1].replace(')', '')
    title = title.split('(')[0]
    # print(website, title, genre)

    return {
        'Family': familyname,
        'Genre': genre,
        'Source': website,
        'Title': title,
    }
```

```python
family_names = ["dofloo", "xorddos", "Darlloz", "lupper", "Gafgyt"]
table = []

for family in family_names:
    filterList = [f for f in html_files if f.find(family)>=0]
    print(family, len(filterList))
    for f in filterList:
        data = parse(f, family)
        table.append(data)

import pandas as pd
pd.DataFrame(table)
```
