```python
def process_eset(origin_text:str):
    origin_text = origin_text.replace('\n',' ')
    # print(origin_text)
    return origin_text

# process_eset(text_body)

content = ""
for i in range(len(reader.pages)):
    page = reader.pages[i]
    parts = []
    page.extract_text(visitor_text=visitor_body)
    raw_text = "".join(parts)
    res = process_eset(raw_text)
    # print(res)
    content += f"{res}\n"

with open('./00_PDFoutput.txt', 'w', encoding='utf-8') as f:
    f.write(content)
```

### Collect APT Groups from Mitre Attack

```python
url = "https://attack.mitre.org/groups/"

from bs4 import BeautifulSoup
import requests

response = requests.get(url)
# response.text
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
colnames = ['ID','Name','Associated Groups','Description']
table = soup.select_one('table')
tbody = table.select_one('tbody')

group_list = set()
rows = tbody.select('tr')
for row in rows:
    this_group = row.select('td')[1].text.strip()
    group_list.add(this_group)
    associated_groups = row.select('td')[2].text.strip()
    if associated_groups != '':
        associated_groups = associated_groups.split(',')
        associated_groups = set(a.strip() for a in associated_groups)
        group_list = group_list | associated_groups
    
    # print(this_group)
    

print(f"rows: {len(rows)}, total groups: {len(group_list)}")
group_list
```
